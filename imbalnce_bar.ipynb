```python
def signed_tick(tick, initial_value=1.0):
    diff = tick['priceclose'] - tick['priceclose'].shift(1)
    return (abs(diff) / diff).ffill().fillna(initial_value)
```

```python
df = pd.read_csv("Future_tick_101T3.csv")
df = df[df['logdate'] >= 20221218]
df = df.drop(['Unnamed: 0'],axis=1)
```
