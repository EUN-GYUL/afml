```python
df['timestamp'] = df.timestamp.map(lambda t: datetime.strptime(t, "%Y%m%d%H%M%S"))
```

```python
time_df = df.set_index('timestamp')
```

```python
def compute_vwap(df):
    
    p = (df['priceclose'] + df['pricelow'] + df['pricehigh'])/3
    q = df['volume']
    
    vwap = np.sum(p * q) / np.sum(q)
    df['vwap'] = vwap
    return df
```

```python
data_time_grp = time_df.groupby(pd.Grouper(freq='15Min'))
num_time_bars = len(data_time_grp)
data_time_vwap = data_time_grp.apply(compute_vwap)
```
